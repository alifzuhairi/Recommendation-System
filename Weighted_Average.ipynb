```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("appstore_games.csv")
df.head()
```

```python
dffinal = df.drop(columns=['URL', 'Subtitle', 'In-app Purchases','Description','Age Rating','Languages','Size','Primary Genre','Genres','Original Release Date','Current Version Release Date'])
```

```python
dffinal.head()
```

```python
v=dffinal['User Rating Count'] 
R=dffinal['Average User Rating'] 
C=dffinal['Average User Rating'].mean()
m=dffinal['User Rating Count'].quantile(0.70) 
```

```python
dffinal['weighted_average']=((R*v)+ (C*m))/(v+m)
```

```python
dffinal.head(5)
```

```python
dfranking=dffinal.sort_values('weighted_average',ascending=False)
```

```python
dfranking.head(10)
```

```python
return dfranking()
```
